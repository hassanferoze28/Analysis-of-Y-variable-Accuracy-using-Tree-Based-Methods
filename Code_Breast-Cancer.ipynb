```python
import numpy as np
import pandas as pd
from sklearn import tree

input_file = "C:/Users/Laiq Ahmed/Desktop/Project 4/breast-cancer-wisconsin.csv"
df = pd.read_csv(input_file, header = 0)
```

```python
#set the dataset into dependent and independent variables and into numpy array
X = df.iloc[:,1:9].values
Y = df.iloc[:,10].values
type(X)

```

```python
#split the data set into 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0 )
```

```python
#scale the data (Feature Scalling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(X_train, Y_train)
dt.score(X_test, Y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cross_val_score(DecisionTreeClassifier(criterion = 'entropy'),X,Y,cv=10)
```

```python
score1 = cross_val_score(DecisionTreeClassifier(criterion = 'entropy'),X,Y,cv=10)
np.average(score1)
```

As we can see above simple Decision tree classifier has a accuracy of 94.285% and with 10 cross folds it has a mean value of 93.98%, Hence Simple decision tree performs better in this case.  

```python
rf = RandomForestClassifier(n_estimators=10)
rf.fit(X_train, Y_train)
rf.score(X_test, Y_test)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10),X,Y,cv=10)
```

```python
Score2 =cross_val_score(RandomForestClassifier(n_estimators=10),X,Y,cv=10)
np.average(Score2)
```

The model Accuracy with Random Forest with 10 fold cross validation is 94.71% Hence it performs better than Decision tree.

```python
from sklearn.neighbors import KNeighborsClassifier  
from sklearn import neighbors
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors = 10)
clf.fit(X_train, Y_train)
clf.score(X_test, Y_test)
```

```python
cross_val_score(KNeighborsClassifier(n_neighbors=10),X,Y,cv=10)
```

```python
Score3 = cross_val_score(KNeighborsClassifier(n_neighbors=3),X,Y,cv=10)
np.average(Score3)
```

The Accuracy Value for k = 10 is 96.714% if we reduce the K value to 3 we can get better results with the accuracy of 97.138%.
